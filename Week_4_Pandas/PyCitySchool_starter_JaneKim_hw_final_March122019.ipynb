```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# The path to CSV files
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#  Calculate the total number of schools
number_of_schools =len(df["school_name"].unique())
number_of_schools
```

```python
# Calculate the total number of students
number_of_students= df["student_name"].count()
number_of_students
```

```python
# Calculate the total budget
total_budget_district= sum(school_data["budget"])
total_budget_district
```

```python
# Calculate the average math score
avg_math_score = df["math_score"].mean()
avg_math_score
```

```python
# Calculate the average reading score
avg_reading_score = df["reading_score"].mean()
avg_reading_score
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = (avg_math_score + avg_reading_score)/2
overall_passing_rate
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)

math_seventy_or_above = df.loc[df["math_score"] >=70, :]
math_seventy_or_above

number_of_students_passing_math = math_seventy_or_above["student_name"].count()
number_of_students_passing_math


percentage_of_students_passing_math_score = (number_of_students_passing_math/number_of_students)*100
percentage_of_students_passing_math_score 
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)

reading_seventy_or_above = df.loc[df["reading_score"] >=70, :]
reading_seventy_or_above

number_of_students_passing_reading = reading_seventy_or_above["student_name"].count()
number_of_students_passing_reading


percentage_of_students_passing_reading_score = (number_of_students_passing_reading/number_of_students)*100
percentage_of_students_passing_reading_score 
```

```python
# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({"Total number of schools": [number_of_schools],
                           "Total number of students": [number_of_students],
                           "Total budget": [total_budget_district],                             
                           "Average math score": [avg_math_score],
                           "Average reading score": [avg_reading_score],
                           "Overall passing rate": [overall_passing_rate],
                           "% Students with a passing math score (>=70)": [percentage_of_students_passing_math_score],
                           "% Students with a passing reading score (>=70)": [percentage_of_students_passing_reading_score]})

# Optional: give the displayed data cleaner formatting
district_summary_df["Total number of schools"] = district_summary_df["Total number of schools"].map("{:,.0f}".format)
district_summary_df["Total budget"] = district_summary_df["Total budget"].map("${:,}".format)
district_summary_df["Total number of students"] = district_summary_df["Total number of students"].map("{:,.0f}".format)
district_summary_df["Average math score"] = district_summary_df["Average math score"].map("{:.2f}".format)
district_summary_df["Average reading score"] = district_summary_df["Average reading score"].map("{:.2f}".format)
district_summary_df["Overall passing rate"] = district_summary_df["Overall passing rate"].map("{:.2f}%".format)
district_summary_df["% Students with a passing math score (>=70)"] = district_summary_df["% Students with a passing math score (>=70)"].map("{:.2f}%".format)
district_summary_df["% Students with a passing reading score (>=70)"] = district_summary_df["% Students with a passing reading score (>=70)"].map("{:.2f}%".format)

district_summary_df
```

## School Summary

# * Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Groupby school data 
groupby_schools= school_data_complete.groupby(["school_name"]) 
groupby_schools.head()

# Get school name and school type 
school_type = school_data.set_index(["school_name"])["type"]
school_type.head()

# Calculate Total number of students 
total_students = groupby_schools["student_name"].count()

# Calculate Total budget 
total_budget = groupby_schools["budget"].sum()
total_budget = groupby_schools["budget"].mean()

# calculate budget per student
per_student_budget = groupby_schools["budget"].mean()/groupby_schools["size"].count()

# calculate average math score
avg_math =groupby_schools["math_score"].mean()

# calculate average reading score
avg_reading =groupby_schools["reading_score"].mean()

# % Passing Math
math_passing_count = math_seventy_or_above["school_name"].value_counts()
math_passing_percent = (math_passing_count/ total_students)*100

# % Passing Reading
reading_passing_count = reading_seventy_or_above["school_name"].value_counts()
reading_passing_percent = (reading_passing_count/ total_students)*100


# Overall passing Rate
overall_passing_rate = (reading_passing_percent + math_passing_percent) /2
```

```python
per_school_summary = pd.DataFrame({"School Type": school_type,
                                  "Total Students": total_students,
                                  "Total School Budget": total_budget,
                                  "Per Student Budget($)": per_student_budget,
                                  "Average Math Score": avg_math,
                                  "Average Reading Score": avg_reading,
                                  "% Passing Math": math_passing_percent,
                                  "% Passing Reading": reading_passing_percent,
                                  "% Overall Passing Rate": overall_passing_rate})

# Mapping /format (these are commented out because formatted numbers were not readable for later calculations )
# per_school_summary["% Passing Math"] = per_school_summary["% Passing Math"].map("{:.2f}%".format)
# per_school_summary["% Passing Reading"] = per_school_summary["% Passing Reading"].map("{:.2f}%".format)
# per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:.2f}".format)
# per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:.2f}".format)
# per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.0f}".format)
# per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.0f}".format)
# per_school_summary["Total Students"] = per_school_summary["Total Students"].map("{:,.0f}".format)
# per_school_summary["% Overall Passing Rate"] = per_school_summary["% Overall Passing Rate"].map("{:.2f}%".format)


per_school_summary.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a series for each grade
grade_9th  = school_data_complete[(school_data_complete["grade"] == "9th")]
grade_10th = school_data_complete[(school_data_complete["grade"] == "10th")]
grade_11th = school_data_complete[(school_data_complete["grade"] == "11th")]
grade_12th = school_data_complete[(school_data_complete["grade"] == "12th")]

# groupby each grades by shool name
grade_9th_group = grade_9th.groupby(["school_name"])
grade_10th_group = grade_10th.groupby(["school_name"])
grade_11th_group = grade_11th.groupby(["school_name"])
grade_12th_group = grade_12th.groupby(["school_name"])

# Get average math score per each grade
grade_9th_math_avg  = pd.DataFrame(grade_9th_group["math_score"].mean())
grade_10th_math_avg = pd.DataFrame(grade_10th_group["math_score"].mean())
grade_11th_math_avg = pd.DataFrame(grade_11th_group["math_score"].mean())
grade_12th_math_avg = pd.DataFrame(grade_12th_group["math_score"].mean())

# Merge each df to build a df that holds all the grades
math_group_9th_10th = pd.merge(grade_9th_math_avg, grade_10th_math_avg, on="school_name")
math_group_9th_10th_11th = pd.merge(math_group_9th_10th, grade_11th_math_avg, on="school_name")
math_group_all_grades = pd.merge(math_group_9th_10th_11th, grade_12th_math_avg, on="school_name")

#  rename column headers
math_group_all_grades.columns= ["9th","10th","11th","12th"]

#  rename index header
math_group_all_grades.index.name = "School Name"

# Optional: give the displayed data cleaner formatting
math_group_all_grades["9th"] = math_group_all_grades["9th"].map("{:.2f}".format)
math_group_all_grades["10th"] = math_group_all_grades["10th"].map("{:.2f}".format)
math_group_all_grades["11th"] = math_group_all_grades["11th"].map("{:.2f}".format)
math_group_all_grades["12th"] = math_group_all_grades["12th"].map("{:.2f}".format)

math_group_all_grades
```
